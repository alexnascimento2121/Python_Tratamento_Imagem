Biblioteca necessaria para manuseio de imagens

````python
import cv2
import matplotlib.pyplot as plt

#carrega as imagem
imagem =cv2.imread('/content/eu.jpg')

#converter a imagem de bgr para rgb
imagem_rgb =cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)

#exibir a imagem
plt.imshow(imagem_rgb)
#plt.axis('off') desabilita os eixos
plt.show()
````

Imagem em cinza-grayscale

````python
#converter a imagem para grayscale
imagem_gray =cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)

#Exibir a imagem com bordas detectadas
plt.imshow(imagem_gray, cmap='gray')
plt.axis('off') #desabilita os eixos
plt.show()
````

Redimensionar Imagem

````python
#redimensionar a imagem
largura =800
altura=600
dimensoes=(largura,altura)
imagem_redimensionada=cv2.resize(imagem,dimensoes,interpolation=cv2.INTER_AREA)

# converter a imagem bgr para rgb
imagem_redimensionada_rgb=cv2.cvtColor(imagem_redimensionada,cv2.COLOR_BGR2RGB)


#exibir a imagem
plt.imshow(imagem_redimensionada_rgb)
plt.axis('off') #desabilita os eixos
plt.show()
````

````python
#converter a imagem para grayscale
imagem_bordas =cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)

#detectar as bordas na imagem
bordas = cv2.Canny(imagem_bordas,100,200)

#Exibir a imagem com bordas detectadas
plt.imshow(bordas, cmap='gray')
plt.axis('off') #desabilita os eixos
plt.show()
````

Redimensionando a imagem para quando for necessario padronizar o tamanho das imagens de entrada

````python
#aplicar suavização com filtro gaussiano
imagem_suavizada = cv2.GaussianBlur(imagem,(15,15),0)
#converter imagem d bgr para rgb
imagem_suavizada_rgb=cv2.cvtColor(imagem_suavizada,cv2.COLOR_BGR2RGB)

#exibir a imagem
plt.imshow(imagem_suavizada_rgb)
plt.axis('off') #desabilita os eixos
plt.show()
````

Desenhar formas na imagem

````python
#definicao de coordenadas e a cor do retangulo
inicio=(500,700) # deslocamento do ini da esq -> dir,deslocamento de cima ->baixo
fim=(100,200)# deslocamento da dir -> esq, deslocamento de baixo ->cima
cor=(255,0,0) #azul em bgr
espessura= 2

#desenhar o retangulo na imagem
imagem_com_retangulo = cv2.rectangle(imagem.copy(),inicio,fim,cor,espessura)

#convertr a imagem para rgb
imagem_com_retangulo_rgb =cv2.cvtColor(imagem_com_retangulo,cv2.COLOR_BGR2RGB)

#exibir a imagem
plt.imshow(imagem_com_retangulo_rgb)
plt.axis('off') #desabilita os eixos
plt.show()
````

Salvando a imagem no disco

````python
#salvar a imagem
caminho_salvar='/content/euprocessado.jpg'
cv2.imwrite(caminho_salvar,imagem_com_retangulo_rgb)
````
